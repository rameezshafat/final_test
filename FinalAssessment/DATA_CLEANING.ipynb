# DATA CLEANING  FOR CSV FILE

```python
dfi['ReleaseYear'] = dfi['ReleaseYear'].fillna(dfi['ReleaseYear'].mean()).astype(int)
dfi
```

```python
dfi['Budget (Crores)'] = dfi['Budget (Crores)'].fillna(dfi['Budget (Crores)'].mean()).astype(int)
dfi
```

```python
dfi['BoxOffice (Crores)'] = dfi['BoxOffice (Crores)'].fillna(dfi['BoxOffice (Crores)'].mean()).astype(int)
dfi
```

```python
dfi['Rating'] = dfi['Rating'].fillna(dfi['Rating'].mean()).round(1)
dfi
```

```python
dfi['Duration (minutes)'] = dfi['Duration (minutes)'].fillna(dfi['Duration (minutes)'].mean()).round(1)
dfi
```

```python
dfi.rename(columns={"Unnamed: 0": "S.No"}, inplace=True)
dfi
```

```python
dfi['Genre'] = dfi['Genre'].str.split().apply(lambda x: ', '.join(x))
dfi
```

```python
dfi[dfi.duplicated('Title', keep=False)]
```

```python
#This syntax will written where there are duplicate entries in Tittle column
dfi = dfi.drop_duplicates(subset='Title', keep='first')
dfi
```

```python
dfi['Title'] = dfi['Title'].str.replace(r'\(.*\)', '', regex=True)
dfi
```

```python
dfi.rename(columns={"S.No": "index"}, inplace=True)
dfi
```

```python
dfi["index"] = range(len(dfi))
dfi
```

# DATA CLEANING OF TOLLYWOOD MOVIES

```python
dft['ReleaseYear'] = dft['ReleaseYear'].fillna(dft['ReleaseYear'].mean()).astype(int)
dft
```

```python
dft['Rating'] = dft['Rating'].fillna(dft['Rating'].mean()).round(1)
dft
```

```python
dft['Duration (minutes)'] = dft['Duration (minutes)'].fillna(dft['Duration (minutes)'].mean()).round(1)
dft
```

```python
dft = dft.drop(dft[dft["MovieID"] == "MOV021"].index)
```

```python
dft.rename(columns={"Unnamed: 0": "S.No"}, inplace=True)
dft
```

```python
dft["Language"] = dft["Language"].fillna("Telugu")
dft
```

```python
dft["S.No"] = range(1, len(dft) + 1)
```

```python
dft = dft.reset_index(drop=True)
```

```python
dft['Genre'] = dft['Genre'].str.split().apply(lambda x: ', '.join(x))
dft
```

```python
dft.rename(columns={"S.No": "index"}, inplace=True)
```

```python
dft["index"] = range(len(dft))
dft
```
